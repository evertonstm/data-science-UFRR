### Como calcular a taxa de mortalidade?

- Pesquise na internet, em local de fonte confiável, como se dá o cálculo da taxa de mortalidade e letalidade.

A taxa de mortalidade e a taxa de letalidade são duas métricas diferentes usadas para avaliar o impacto de uma doença como a COVID-19. Aqui está a diferença entre elas e como são calculadas:

1. **Taxa de Mortalidade**: A taxa de mortalidade é a proporção de pessoas que morreram de uma doença em relação ao número total de casos diagnosticados. Para calcular a taxa de mortalidade da COVID-19, você divide o número total de mortes pelo número total de casos confirmados e multiplica por 100 para obter a porcentagem.

   $$ \text{Taxa de Mortalidade} = \frac{\text{Número total de mortes}}{\text{Número total de casos confirmados}} \times 100 $$

2. **Taxa de Letalidade**: A taxa de letalidade é a proporção de pessoas que morreram de uma doença em relação ao número total de pessoas infectadas, incluindo os casos assintomáticos ou não detectados. Para calcular a taxa de letalidade da COVID-19, você divide o número total de mortes pelo número total de pessoas infectadas (não apenas os casos confirmados) e multiplica por 100 para obter a porcentagem.

   $$ \text{Taxa de Letalidade} = \frac{\text{Número total de mortes}}{\text{Número total de pessoas infectadas}} \times 100 $$

Essas duas métricas são importantes para entender o impacto de uma doença, mas elas podem produzir resultados diferentes dependendo da disponibilidade de testes e da precisão dos dados de contagem de casos e mortes. A taxa de mortalidade é mais comumente usada porque os dados sobre casos confirmados são mais disponíveis e precisos em comparação com o número total de pessoas infectadas.
